```python
import fitz
import pandas as pd
import numpy as np
import openpyxl
```

```python
rod="Keppel Corporation Limited Report to Shareholders 2018"
def cleaning(row):
    try:
        if int(row):
            return np.nan
    except:
        if str(row).startswith('<'):
            return np.nan
        elif str(row).startswith(rod):
            return np.nan

        else:
            return row

def get_paras(path,pno):
    doc= fitz.open(path)
    page = doc.load_page(pno)
    text = page.get_text("blocks")
    cols=[]
    sets=set()
    for i in text:
        sets.add(i[0])
    sets=list(sets)
    sets.sort()
    res=[]
    #print(sets)
    for xcoord in sets:
        temp=[]
        for i in text:
            if xcoord == i[0]:
                temp.append(i[4])
        res.extend(temp)
    #print(res)
    df=pd.DataFrame(res,columns=["rows"])
    #print(df)
    df["rows"]=df["rows"].str.replace("\n","")  
    df['rows']=df['rows'].apply(cleaning)
    df.dropna(inplace=True)
    df.reset_index(inplace=True)
    df.drop(columns=['index'],inplace=True)
    df.to_excel("output.xlsx")
    return df
```

```python
path=input("Enter the filename")
pgn=int(input("Enter the pagenumber"))
path="keppel-corporation-limited-annual-report-2018.pdf"
lol=get_paras(path,pgn-1)
lol
```
